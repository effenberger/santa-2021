```python
# this returns true if s is a permutation of 1,2,3,4,5,6,7
def is_perm(s):
    y = True
    for k in range(1,8):
        y = y&(str(k) in s)
        if not y: break
    return y
```

```python
permutations = []

# go through the superperm string and look at chunks with 7 chars
# check if it is a permutation, that it is not already counted and exclude permutations starting with '54'
# perms with '54' are the ones that need to be in every of the three strings
for k in range(len(best)-6):
    s = best[k:k+7]
    if (is_perm(s))&(s not in permutations)&(s[:2]!='54'):
        permutations.append(s)
        
print('There are',len(permutations),'that can be split among 3 strings')
```

```python
def hamming_distance(str1, str2):
    return sum( (c1!=c2) for c1, c2 in zip(str1, str2))

def offset(s1, s2):
    assert(len(s1)==len(s2))
    ln = len(s1)
    j = ln
    for k in range(0,ln):
        if hamming_distance(s1[k:],s2[:7-k])==0:
            j=k
            break
    return j
```

```python
def get_tsp_solution(group, folder):
    
    # CREATE DISTANCE MATRIX
    SIZE = len(group)
    M = np.zeros((SIZE, SIZE), dtype='int8')
    for j in range(SIZE):
        #if j%25==0: print(j,', ',end='')
        for k in range(SIZE):
            M[j,k] = offset(group[j],group[k])
            
    os.mkdir(f"./lkh/{folder}")
            
    # WRITE PROBLEM FILE
    f = open("./lkh/" + folder + "/group.par", 'w')
    f.write("PROBLEM_FILE = ../lkh/" + folder + "/distances.atsp\n")
    f.write("TOUR_FILE = ../lkh/" + folder + "/output.txt\n")
    f.write(f"OPTIMUM = {SIZE}\n")
    f.write("MOVE_TYPE = 5\n")
    f.write("PATCHING_C = 3\n")
    f.write("PATCHING_A = 2\n")
    f.write("RUNS = 1\n")
    f.write("TIME_LIMIT = 20\n") #seconds
    f.close()
    
    # WRITE PARAMETER FILE
    f = open("./lkh/" + folder + "/distances.atsp",'w')
    f.write("NAME: distances\n")
    f.write("TYPE: ATSP\n")
    f.write("COMMENT: Asymmetric TSP\n")
    f.write(f"DIMENSION: {SIZE}\n")
    f.write("EDGE_WEIGHT_TYPE: EXPLICIT\n")
    f.write("EDGE_WEIGHT_FORMAT: FULL_MATRIX\n")
    f.write("EDGE_WEIGHT_SECTION\n")
    for j in range(SIZE):
        #if j%25==0: print(j,', ',end='')
        for k in range(SIZE):
            f.write(f"{M[j,k]:2d} ") 
        f.write("\n")
    f.close()
    
    # EXECUTE TSP SOLVER
    !cd LKH-3.0.7; ./LKH ../lkh/{folder}/group.par
    
    # READ RESULTING ORDER
    with open("./lkh/" + folder + "/output.txt") as f:
        lines = f.readlines()
    for i,ln in enumerate(lines):
        if 'TOUR_SECTION' in ln: break
    perms = [int(x[:-1]) for x in lines[i+1:-2] ]
    
    # CREATE STRING
    result = group[ perms[0]-1 ]
    for k in range(1,len(perms)):
        s1 = group[ perms[k-1]-1 ]
        s2 = group[ perms[k]-1 ]
        d = offset(s1,s2)
        assert(d!=0)
        result += s2[-d:]
    
    return result
```

```python
def get_tsp_submission(t1, t2, adjust1, adjust2, folder):
    group1 = permutations[:t1-adjust1] + mandatory
    group2 = permutations[t1-adjust1:t2+adjust2] + mandatory
    group3 = permutations[t2+adjust2:] + mandatory
    
    string1 = get_tsp_solution(group1, folder + "_g1")
    string2 = get_tsp_solution(group2, folder + "_g2")
    string3 = get_tsp_solution(group3, folder + "_g3")
    
    lengths = sorted([len(string1), len(string2), len(string3)], reverse=True)
    
    # CONVERT NUMBERS TO EMOJIS
    replace_dict = {
     '5': '🎅',
     '4': '🤶',
     '8': '🌟',
     '1': '🦌',
     '2': '🧝',
     '3': '🎄',
     '6': '🎁',
     '7': '🎀'
    }

    for k,v in replace_dict.items():
        string1 = string1.replace(k, v)
        string2 = string2.replace(k, v)
        string3 = string3.replace(k, v)
        
    
    sub = pd.DataFrame()
    sub['schedule'] = [string1, string2, string3]
    
    ts = time.time()
    sub.to_csv('submission_' + folder + '_' + '_'.join([str(x) for x in lengths]) + "_" + str(round(ts)), index=False)
```
